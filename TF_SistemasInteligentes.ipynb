```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

# Supongamos que 'uploaded' contiene tus datos (reemplázalo con tu método de carga)
uploaded = files.upload()

try:
    # Intenta leer con delimitador de punto y coma primero
    df = pd.read_csv(next(iter(uploaded)), encoding='latin1', sep=';')
except UnicodeDecodeError:
    df = pd.read_csv(next(iter(uploaded)), encoding='iso-8859-1', sep=';')

# Eliminar cualquier columna no numérica si es necesario
df_numeric = df.select_dtypes(include=['number'])

# Instanciar el modelo KMeans (elige el número de clusters que necesites)
kmeans = KMeans(n_clusters=3)

# Ajustar el modelo a los datos numéricos
kmeans.fit(df_numeric)

# Obtener los centroides de los clusters
centroides = kmeans.cluster_centers_

# Predecir los clusters para cada punto de datos
clusters = kmeans.predict(df_numeric)

# Añadir la columna 'cluster' al DataFrame original
df['cluster'] = clusters

# Guardar el DataFrame con la columna 'cluster' (opcional)
# df.to_csv(f'{archivo.split(".")[0]}_clusters.csv', index=False)

# Visualizar los centroides y los puntos de datos
# (Asegúrate de seleccionar columnas numéricas para la visualización)
plt.scatter(df_numeric.iloc[:, 0], df_numeric.iloc[:, 1], c=clusters, cmap='viridis')
plt.scatter(centroides[:, 0], centroides[:, 1], marker='x', s=100, c='red', label='Centroides')
plt.title('Clustering KMeans')
plt.xlabel('Característica 1')  # Reemplaza con el nombre real de tu columna
plt.ylabel('Característica 2')  # Reemplaza con el nombre real de tu columna
plt.legend()
plt.show()

# Explicación del gráfico
print("Explicación del gráfico:")
print("Cada punto representa una observación en tus datos.")
print("El color de cada punto indica el cluster al que pertenece.")
print("Las 'x' rojas representan los centroides de cada cluster.")
print("Los centroides son los 'centros' de cada grupo de puntos.")
print("Este gráfico te ayuda a visualizar cómo KMeans ha agrupado tus datos.")
```
