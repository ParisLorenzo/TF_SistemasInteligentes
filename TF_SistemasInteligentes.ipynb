```python
import pandas as pd

df1 = pd.read_csv('SI-TF(Listado instituciones CEBE).csv', encoding='latin1')
df2 = pd.read_csv('SI-TF(Estudiantes en EBE).csv', encoding='latin1')
df3 = pd.read_csv('SI-TF(Estudiantes disc.csv', encoding='latin1')

# Muestra del dataset
df1.head()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# Selecciona las columnas numéricas relevantes para el clustering
numeric_columns = ['Alumnos (2015)', 'Docentes (2015)', 'Secciones (2015)']
numeric_data = df1[numeric_columns]

# Imputar valores faltantes (si los hay)
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(numeric_data)

# Escalar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(imputed_data)

# Aplicar KMeans
kmeans = KMeans(n_clusters=3, n_init=10)  # Ajusta el número de clusters según sea necesario
kmeans.fit(scaled_data)
clusters = kmeans.labels_

# Agregar los clusters a los datos originales
df1['Cluster'] = clusters

# Visualización de los resultados del clustering
plt.scatter(scaled_data[:, 0], scaled_data[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Alumnos (2015)')
plt.ylabel('Docentes (2015)')
plt.title('Clustering Results')
plt.show()

# Asegurarse de que solo las columnas numéricas se usen para el análisis de clusters
numeric_df = df1[numeric_columns].copy()

# Añadir la columna de clusters para análisis
numeric_df.loc[:, 'Cluster'] = clusters

# Análisis de puntos de datos por cluster
cluster_summary = numeric_df.groupby('Cluster').mean()
print("Resumen de Clusters:")
print(cluster_summary)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
import seaborn as sns

# Seleccionar las columnas numéricas relevantes para el clustering
disability_columns = [
    'Estudiantes con discapacidad Intelectual (EBE y EBR)',
    'Estudiantes con discapacidad Auditiva(EBE y EBR)',
    'Estudiantes con discapacidad Visual(EBE y EBR)',
    'Estudiantes con discapacidad Sordoceguera(EBE y EBR)',
    'Estudiantes con discapacidad Motora(EBE y EBR)',
    'Estudiantes con discapacidad TEA(EBE y EBR)',
    'Estudiantes con discapacidad Otra NEE (EBE y EBR)'
]

# Calcular el total de estudiantes con alguna discapacidad
df3['Total_Estudiantes_Discapacidades'] = df3[disability_columns].sum(axis=1)

# Calcular el porcentaje de estudiantes con cada tipo de discapacidad
for column in disability_columns:
    df3[column + '_Pct'] = df3[column] / df3['Total_Estudiantes_Discapacidades'] * 100

# Seleccionar las columnas de porcentaje para el clustering
percentage_columns = [col + '_Pct' for col in disability_columns]
numeric_data = df3[percentage_columns]

# Imputar valores faltantes (si los hay)
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(numeric_data)

# Escalar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(imputed_data)

# Aplicar KMeans
kmeans = KMeans(n_clusters=4, n_init=10)  # Ajusta el número de clusters según sea necesario
kmeans.fit(scaled_data)
clusters = kmeans.labels_

# Agregar los clusters a los datos originales
df3['Cluster'] = clusters

# Visualización de los resultados del clustering
plt.figure(figsize=(10, 8))
sns.heatmap(df3.groupby('Cluster')[percentage_columns].mean().T, annot=True, cmap='YlGnBu')
plt.title('Heatmap de la Media de Porcentajes de Estudiantes con Discapacidades por Cluster')
plt.show()

# Análisis de puntos de datos por cluster
numeric_df = df3[percentage_columns].copy()
numeric_df['Cluster'] = clusters
cluster_summary = numeric_df.groupby('Cluster').mean()
print("Resumen de Clusters:")
print(cluster_summary)

# Mostrar la distribución de clusters por regiones
region_distribution = df3.groupby(['Region', 'Cluster']).size().unstack(fill_value=0)
print("\nDistribución de Clusters por Región:")
print(region_distribution)
```
